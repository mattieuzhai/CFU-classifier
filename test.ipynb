```python
from PIL import Image
import os
```

```python
def resize_images(input_dir, output_dir, size):
    """
    Resize all images in input_dir to the given size and save to output_dir.

    Args:
        input_dir (str): Path to folder containing source images.
        output_dir (str): Path to folder where resized images will be saved.
        size (tuple): Target size as (width, height) in pixels.
    """
    os.makedirs(output_dir, exist_ok=True)
    supported = {".jpg", ".jpeg", ".png", ".tif", ".tiff", ".bmp"}

    for filename in os.listdir(input_dir):
        ext = os.path.splitext(filename)[1].lower()
        if ext not in supported:
            continue

        src_path = os.path.join(input_dir, filename)
        dst_path = os.path.join(output_dir, filename)

        with Image.open(src_path) as img:
            resized = img.resize(size, Image.LANCZOS)
            resized.save(dst_path)
            print(f"Saved {filename} -> {size}")

    print("Done.")
```

```python
import shutil

def copy_images_without_tag(input_dir, output_dir, tag):
    """
    Recursively finds all leaf directories under input_dir and copies any image
    whose filename does NOT contain `tag` to output_dir (flat, no subfolders).

    Args:
        input_dir (str): Root directory to search.
        output_dir (str): Destination directory (created if it doesn't exist).
        tag (str): Substring to exclude — files containing this are skipped.
    """
    supported = {".jpg", ".jpeg", ".png", ".tif", ".tiff", ".bmp"}
    os.makedirs(output_dir, exist_ok=True)

    for dirpath, dirnames, filenames in os.walk(input_dir):
        # Only process leaf directories (no subdirectories)
        if dirnames:
            continue

        for filename in filenames:
            ext = os.path.splitext(filename)[1].lower()
            if ext not in supported:
                continue
            if tag in filename:
                continue

            src = os.path.join(dirpath, filename)
            dst = os.path.join(output_dir, filename)

            shutil.copy2(src, dst)
            print(f"Copied: {filename}")

    print("Done.")
```

```python
# Example usage — edit these paths and size before running
resize_images(
    input_dir="data/images_without_mask",
    output_dir="data/images_resized",
    size=(6400, 6400),
)
```

```python
from sklearn.model_selection import train_test_split

def split_dataset(input_dir, output_dir, train_ratio=0.7, val_ratio=0.15, test_ratio=0.15, random_state=42):
    """
    Splits images in input_dir into train/val/test sets and copies them into
    the Ultralytics-compatible structure: output_dir/{split}/images/

    Args:
        input_dir (str): Flat directory of images to split.
        output_dir (str): Destination root (subfolders created automatically).
        train_ratio (float): Proportion for training set.
        val_ratio (float): Proportion for validation set.
        test_ratio (float): Proportion for test set.
        random_state (int): Seed for reproducibility.
    """
    assert abs(train_ratio + val_ratio + test_ratio - 1.0) < 1e-6, "Ratios must sum to 1"

    supported = {".jpg", ".jpeg", ".png", ".tif", ".tiff", ".bmp"}
    images = [f for f in os.listdir(input_dir) if os.path.splitext(f)[1].lower() in supported]

    # First split: train vs (val + test)
    train_files, val_test_files = train_test_split(
        images, test_size=(val_ratio + test_ratio), random_state=random_state
    )

    # Second split: val vs test
    val_files, test_files = train_test_split(
        val_test_files, test_size=(test_ratio / (val_ratio + test_ratio)), random_state=random_state
    )

    for split, files in [("train", train_files), ("val", val_files), ("test", test_files)]:
        images_dir = os.path.join(output_dir, split, "images")
        os.makedirs(images_dir, exist_ok=True)
        os.makedirs(os.path.join(output_dir, split, "labels"), exist_ok=True)
        for filename in files:
            shutil.copy2(os.path.join(input_dir, filename), os.path.join(images_dir, filename))
        print(f"{split}: {len(files)} images")

    print("Done.")
```

```python
# Example usage — edit before running
split_dataset(
    input_dir="data/images_resized",
    output_dir="data/split",
    train_ratio=0.33,
    val_ratio=0.34,
    test_ratio=0.33,
)
```
